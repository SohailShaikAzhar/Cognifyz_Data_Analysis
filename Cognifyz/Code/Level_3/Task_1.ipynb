###  Analyze the text reviews to identify the most common positive and negative keywords.

```python
import pandas as pd
from collections import Counter

df = pd.read_csv("Dataset.csv")

df["Rating text"] = df["Rating text"].astype(str).str.lower()

positive_words = ["excellent", "amazing", "good", "great", "best", "fantastic", "awesome"]
negative_words = ["poor", "bad", "terrible", "worst", "awful", "disappointing"]

positive_count = Counter()
negative_count = Counter()

for text in df["Rating text"]:
    words = text.split()
    
    for word in words:
        if word in positive_words:
            positive_count[word] += 1

    for word in words:
        if word in negative_words:
            negative_count[word] += 1

print("Most common positive words:", positive_count)
print("Most common negative words:", negative_count)

```

### Calculate the average length of reviews and explore if there is a relationship between review length and rating.


```python
import matplotlib.pyplot as plt

df['Rating text'] = df['Rating text'].astype(str)

df['review_length'] = df['Rating text'].apply(lambda x: len(x.split()))

average_length = df['review_length'].mean()
print(f"Average review length: {average_length:.2f} words")

plt.figure(figsize=(10, 6))
plt.scatter(df['Aggregate rating'], df['review_length'], alpha=0.5, color='teal')
plt.xlabel("Aggregate Rating")
plt.ylabel("Review Length (words)")
plt.title("Relationship between Review Length and Aggregate Rating")
plt.grid(True)
plt.show()

correlation = df['Aggregate rating'].corr(df['review_length'])
print(f"Correlation between aggregate rating and review length: {correlation:.2f}")
```
